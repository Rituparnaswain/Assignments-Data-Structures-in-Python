```python
#  np.array()
# Purpose: This function is used to create a new NumPy array from the input data. It is the most general function for array creation.

# np.asarray()
# Purpose: This function is used to convert the input to a NumPy array if it is not already one. It avoids making unnecessary copies of the data.

# np.asanyarray()
#Purpose: This function is similar to np.asarray() but with an additional feature: it preserves subclasses of ndarray. This is useful when dealing with custom array types that inherit from ndarray.
```

```python
# The difference between Deep copy and shallow copyX
```

```python
# Shallow Copy

# A shallow copy of an object creates a new object, but does not create copies
# of nested objects (objects that are referenced by the original object).
# Instead, it copies the references to these objects. Therefore, changes to
# nested objects in the copied object will reflect in the original object,
# because both copies share references to the same nested objects.
```

```python
# Deep Copy

# A deep copy of an object creates a new object and recursively copies all
# objects found in the original. This means that the new object and its nested
# objects are entirely independent of the original object.
```

```python
# 17 Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
# origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

x = np.random.randn(100)
y = np.random.randn(100)


quadrants = np.zeros(len(x), dtype=int)
quadrants[(x >= 0) & (y >= 0)] = 1
quadrants[(x < 0) & (y >= 0)] = 2
quadrants[(x < 0) & (y < 0)] = 3
quadrants[(x >= 0) & (y < 0)] = 4


data = {'x': x, 'y': y, 'Quadrant': quadrants}
df = pd.DataFrame(data)


sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='viridis')


plt.xlabel('X-axis')
plt.ylabel('Y-axis')


plt.title('Quadrant-wise Scatter Plot')


plt.legend(title='Quadrant')

plt.grid(True)
plt.show()

```

```python
# 18 With Bokeh, plot a line chart of a sine wave function, add grid lines,
# label the axes, and set the title as 'Sine Wave Function'
```

```python
from bokeh.plotting import figure, output_notebook, show
import numpy as np

output_notebook()

x = np.linspace(0, 4*np.pi, 100)

y = np.sin(x)

p = figure(title='Sine Wave Function', x_axis_label='X-axis', y_axis_label='Y-axis')

p.line(x, y, line_width=2)

p.grid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.5

show(p)
```

```python
# 19 Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
# values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
# Bar Chart'
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 10, size=len(categories))

p = figure(x_range=categories, height=350, title='Random Categorical Bar Chart',
           x_axis_label='Categories', y_axis_label='Values')

source = ColumnDataSource(data=dict(categories=categories, values=values, color=values))

bars = p.vbar(x='categories', top='values', width=0.9, color='color', legend_field='categories', source=source)

hover = HoverTool()
hover.tooltips = [('Category', '@categories'), ('Value', '@values')]
p.add_tools(hover)


p.xaxis.major_label_orientation = 1.2

show(p)
```
