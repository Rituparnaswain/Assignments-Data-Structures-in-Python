```python
# 5.Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate the synthetic dataset with correlated features
np.random.seed(42)  # For reproducibility

# Generate 100 samples
n_samples = 100

# Generate random data for the first feature
feature_1 = np.random.normal(0, 1, n_samples)

# Create additional features that are correlated with the first feature
feature_2 = feature_1 * 0.8 + np.random.normal(0, 0.2, n_samples)
feature_3 = feature_1 * -0.5 + np.random.normal(0, 0.3, n_samples)
feature_4 = feature_1 * 0.6 + feature_2 * 0.2 + np.random.normal(0, 0.1, n_samples)
feature_5 = feature_2 * -0.7 + feature_3 * 0.4 + np.random.normal(0, 0.2, n_samples)

# Combine into a DataFrame
data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4,
    'Feature 5': feature_5
})

# Step 2: Calculate the correlation matrix
correlation_matrix = data.corr()

# Step 3: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Adding titles and labels
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(25)

# Dataset creation
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)

# Bubble chart creation
plt.figure(figsize=(10, 6))

# Create a bubble chart
plt.scatter(
    x=df['GDP'],
    y=df['Population'],
    s=df['Population'],  # Bubble size proportional to the population
    alpha=0.5,
    c=np.random.rand(5)  # Random colors for each bubble
)

# Adding titles and labels
plt.title('Country Population vs GDP')
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')

# Annotate each point with the country name
for i in range(len(df)):
    plt.text(
        df['GDP'][i],
        df['Population'][i],
        df['Country'][i],
        fontsize=9,
        ha='right'
    )

# Show the plot
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Ensure the plot displays in the notebook
output_notebook()

# Create x values from 0 to 10
x = np.linspace(0, 10, 500)
# Compute y values as the sine of x
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Sample data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create the Bokeh histogram
output_notebook()
p = figure(title="Histogram of Random Data", background_fill_color="#fafafa")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper
from bokeh.transform import linear_cmap
from bokeh.io import output_notebook

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Set up Bokeh
output_notebook()

# Create a color mapper
colors = ['#440154', '#404387', '#29788E', '#22A784', '#79D151', '#FDE724']
mapper = linear_cmap(field_name='value', palette=colors, low=data_heatmap.min(), high=data_heatmap.max())

# Create the heatmap plot
p = figure(title="Heatmap", x_range=(0, 1), y_range=(0, 1), toolbar_location=None)
p.rect(x=xx.flatten(), y=yy.flatten(), width=1/10, height=1/10,
       fill_color=mapper, line_color=None, alpha=0.6)

# Show the plot
show(p)
```
